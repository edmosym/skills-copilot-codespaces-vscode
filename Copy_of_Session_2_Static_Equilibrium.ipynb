## Multiple Loads on an Axial Bar
Multiple Loads on an Axial Bar.png

```python
F_A=100 # (N)
F_B=20 # (N)
F_C=50 # (N)
F_D=F_A-F_B-F_C
print('F_D=', np.round(F_D,1), 'N')

F_AB=F_A
F_BC=F_A-F_B
F_CD=F_D

print('F_AB=', np.round(F_AB,1),'N; F_BC=',np.round(F_BC,1),'N; F_CD=', np.round(F_CD,1),'N')
```

## Static Equilibrium for Beam Bending
Static Equilibrium for Beam Bending.png

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
L=1
P=100
x=np.linspace(0,L,num=50)

V=P+x*0
M=P*(x-L)

fig=plt.figure()
ax=fig.add_subplot(211)
plt.plot(x,V,'bo-')
ax.set_xlim([0, L])
ax.set_xlabel('x (m)',size=20)
ax.set_ylabel('V (N)',size=20)
ax.set_title('Shear Force and Bending Moment $L=$ '+str(np.round(L))+' m and $P=$'+str(np.round(P))+' N');

ax=fig.add_subplot(212)
plt.plot(x,M,'bo-')
ax.set_xlim([0, L])
ax.set_xlabel('x (m)',size=20)
ax.set_ylabel('M (Nm)',size=20);
```

# Method of Joints for Trusses
Method of Joints for Trusses.png

```python
P=10 # (N)
A=np.array([[1,0,np.sqrt(2)/2],[0,1,-np.sqrt(2)/2],[0,1,0]])
display(A)
B=np.array([P,0,P])
display(B)
X=linalg.solve(A,B)
display(X)
```
