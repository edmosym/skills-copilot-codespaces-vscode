# Session 3: Axial Loading

````python
import numpy as np
import sympy as sp
import matplotlib
import matplotlib.pyplot as plt
````

Appropriate Spring Coefficients for Bars

````python
D=0.01 # Diameter (m)
L=0.1 # Length (m)
E=210e9 # Elastic modulus for steel (Pa)
A=np.pi*D**2/4
k=E*A/L
print('k=', np.round(k*1e-6,1), 'kN/mm')
````

````python
D_array=np.linspace(0.001,0.02,100)
k_array=E*np.pi*D_array**2/4*1/L

fig=plt.figure()
ax=fig.add_subplot(111)
plt.plot(D_array,k_array*1e-6,'bo-')
ax.set_xlim([D_array[0], D_array[-1]])
ax.set_xlabel('D (m)',size=20)
ax.set_ylabel('k (kN/mm)',size=20)
ax.set_title('Spring coefficient k for $L=$ '+str(np.round(L,2))+' m and $E=$'+str(np.round(E*1e-9,0))+' GPa');
````

## Example 3.1

## Example 3.2 (with values)
image.png

````python
k1=200e6 #N/m
k2=150e6 #N/m
k3=100e6 #N/m
F=1000 #N
keq=(2*k1+3*k2)*k3/(2*k1+3*k2+3*k3)

delta_3pt2=F/keq # Displacement with no rotation (m))
print('keq=',np.round(keq*1e-6,1),'kN/mm')
print('delta for Example 3.2=',np.round(delta_3pt2*1e6,1),'um')
````

## Example 3.3 with values
image.png

````python
#Same values as from above
k1=200e6 #N/m
k2=150e6 #N/m
k3=100e6 #N/m
F=1000 #N
# Some lengths for lower beam
L_AC=1 # (m)
L_AB=0.3 # (m)
L_BC=L_AC-L_AB

F1=L_BC/L_AC*F
print('Force F1=',np.round(F1,1),'N')
F2=L_AB/L_AC*F
print('Force F2=',np.round(F2,2),'N')
delta1=F1*3/2/k1
print('delta1=',np.round(delta1*1e6,3),'um')
delta2=F2/k2
print('delta2=',np.round(delta2*1e6,3),'um')
print('delta1-delta2=',np.round((delta1-delta2)*1e6,3),'um')

theta=F/L_AC**2*(3*L_BC/(2*k1)-L_AB/k2)
print('Angle of rotation =',np.round(theta*180/np.pi*1e6,1),'micro deg')

delta_3pt3=L_AB/(k2*L_AC)*F+L_BC*theta+F/k3
print('delta for Example 3.3=',np.round(delta_3pt3*1e6,1),'um')
````

## Example 3.4 Superposition and Axial Loading with Values
image.png

````python
k1=200e6 # N/m
k2=150e6 # N/m
k3=100e6 # N/m
F1=1000 # N
F2=500 # N
F3=1500 # N
delta_3pt4=F1/k1-F2*(1/k1+1/k2)+F3*(1/k1+1/k2+1/k3)

print('delta for Example 3.4 =',np.round(delta_3pt4*1e6,1),'um')
````

## Example 3.5 Indeterminate Loading with Values

image.png

````python
k1=200e6 # N/m
k2=100e6 # N/m
F=1000 # N
delta1_3pt5=F/k1
R1=k1/(k1+k2)*F
F_R=k2/(k1+k2)*F

print('delta for Example 3.5 =',np.round(delta1_3pt5*1e6,1),'um')
print('R1 =',np.round(R1,1),'N')
print('F_R =',np.round(F_R,1),'N')
````
